```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler,LabelEncoder
```

```python
# Select only numerical features for scaling .
numeric_feature=data.select_dtypes(include=['float64','int64']).columns
```

```python
# Standard Scaling .
standard_scale=StandardScaler()
data_standard_scale=data.copy()
data_standard_scale[numeric_feature]=standard_scale.fit_transform(data[numeric_feature])

# Display the first few rows of the standard scaled dataframe .
print('\nSTANDARD SCALED DATAFRAME :')
print(data_standard_scale.head())
```

```python
# Min-Max Scaling .
min_max=MinMaxScaler()
data_min_max=data.copy()
data_min_max[numeric_feature]=min_max.fit_transform(data[numeric_feature])

# Display the first few rows of the min-max scaled dataframe .
print('\nMIN-MAX SCALED DATAFRAME :')
print(data_min_max.head())
```

The scenarios where each scaling technique is preferred and why.

**•	Standard Scaling** is generally preferred when features are normally distributed, when you need to standardize the scale of features with different units,or when outliers are not a significant issue .

**• Min-Max Scaling** is ideal when features need to be transformed to a specific range, especially for algorithms sensitive to feature scaling or when maintaining relative distances between data points is crucial .

**2. Encoding Techniques:**

```python
cat_cols=data.select_dtypes(include=['object']).columns
# Applying One-Hot Encoding to categorical variables with less than 5 categories .
for col in cat_cols:
  if data[col].nunique()<5:
    data=get_dummies(data,columns=[col],prefix=[col],drop_first=True)
# Display the first few rows of the dataframe after one-hot encoding .
print('\nDATAFRAME AFTER ONE-HOT ENCODING :')
print(data.head())
```

```python
# Select categorical columns .
cat_to_encode=data.select_dtypes(include=['object']).columns
for col in cat_to_encode:
  # Check number of unique categories .
  lab_encoder=LabelEncoder()
  data[col]=lab_encoder.fit_transform(data[col])

# Display the dataframe after label encoding .
print('\nDATAFRAME AFTER LABEL ENCODING :')
print(data.head())
```

```python
data
```

Discuss the pros and cons of One-Hot Encoding and Label Encoding.

**•	One-Hot Encoding** is generally preferred for nominal (categorical without intrinsic order) data because it avoids implying any order between categories and is compatible with many algorithms. However, it can lead to high dimensionality and inefficiencies if the number of categories is large .

**• Label Encoding** is more compact and can be appropriate for ordinal data where there is a meaningful order among categories. However, it should be avoided for nominal data as it can introduce misleading ordinal relationships .

**4. Feature Selection:**

```python
from sklearn.ensemble import IsolationForest
```

```python
# Identifying outliers using isolation forest .
clf=IsolationForest(contamination=0.05,random_state=42)
outlier_preds=clf.fit_predict(data.dropna())
# Creating a mask to identify outliers .
outlier_mask=outlier_preds==-1
# Remove outliers from the dataframe .
df_no_outliers=data.loc[~outlier_mask]
# Display the datarame after removing outliers .
print('\nDATAFRAME AFTER REMOVING OUTLIERS :')
df_no_outliers
```
